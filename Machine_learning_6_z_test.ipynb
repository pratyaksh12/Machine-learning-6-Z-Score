```python
import statistics as st
import pandas as pd
import plotly.figure_factory as ff
import plotly.graph_objects as go
import random

df = pd.read_csv("School2.csv")
population_data = df['Math_score'].tolist()


mean_population = st.mean(population_data)
fig = ff.create_distplot([population_data],['distribution of population data'],show_hist = False)
fig.add_trace(go.Scatter(x = [mean_population,mean_population],y = [0,0.013]))
fig.show()
```

```python
def sampling_distribution(counter,data):

  mean_list = []
  for i in range(counter):
    random_index = random.randint(0,len(data)-1)
    value = data[random_index]
    mean_list.append(value)
  mean = st.mean(mean_list)
  return mean

sampling_population_data = []
for i in range(1200):
  mean = sampling_distribution(100,population_data)
  sampling_population_data.append(mean)

sampling_population_mean = st.mean(sampling_population_data)
sampling_population_stdev = st.stdev(sampling_population_data)

fig = ff.create_distplot([sampling_population_data],['sampling distribution of the population data'],show_hist = False)
fig.add_trace(go.Scatter(x = [sampling_population_mean,sampling_population_mean],y = [0,0.15],mode = 'lines',name = 'mean of sampling population distribution'))
fig.show()


```

```python
sampling_sample_data =[]
for i in range(200):
  mean = sampling_distribution(100,sample_data)
  sampling_sample_data.append(mean)
sample_mean = st.mean(sampling_sample_data)
fig = ff.create_distplot([sampling_sample_data],['distribution of sample data'],show_hist = False)
fig.add_trace(go.Scatter(x = [sample_mean,sample_mean],y = [0,0.15],mode = 'lines',name = 'mean of sample data'))
fig.show()
```

```python
sampling_stdev_population_1_start,sampling_stdev_population_1_end = sampling_population_mean - sampling_population_stdev,sampling_population_mean + sampling_population_stdev
sampling_stdev_population_2_start,sampling_stdev_population_2_end = sampling_population_mean - (2*sampling_population_stdev),sampling_population_mean + (2*sampling_population_stdev)
sampling_stdev_population_3_start,sampling_stdev_population_3_end = sampling_population_mean - (3*sampling_population_stdev),sampling_population_mean + (3*sampling_population_stdev)
```

```python
fig = ff.create_distplot([sampling_population_data],['sampling distribution of the population data'],show_hist = False)
fig.add_trace(go.Scatter(x = [sampling_population_mean,sampling_population_mean],y = [0,0.15],mode = 'lines',name = 'mean population distribution'))
fig.add_trace(go.Scatter(x = [sampling_stdev_population_1_start,sampling_stdev_population_1_start],y = [0,0.15],mode = 'lines',name = 'standard deviation 1 start'))
fig.add_trace(go.Scatter(x = [sampling_stdev_population_1_end,sampling_stdev_population_1_end],y = [0,0.15],mode = 'lines',name = 'standard deviation 1 end'))
fig.add_trace(go.Scatter(x = [sampling_stdev_population_2_start,sampling_stdev_population_2_start],y = [0,0.15],mode = 'lines',name = 'standard deviation 2 start'))
fig.add_trace(go.Scatter(x = [sampling_stdev_population_2_end,sampling_stdev_population_2_end],y = [0,0.15],mode = 'lines',name = 'standard deviation 2 end'))
fig.add_trace(go.Scatter(x = [sampling_stdev_population_3_start,sampling_stdev_population_3_start],y = [0,0.15],mode = 'lines',name = 'standard deviation 3 start'))
fig.add_trace(go.Scatter(x = [sampling_stdev_population_3_end,sampling_stdev_population_3_end],y = [0,0.15],mode = 'lines',name = 'standard deviation 3 end'))
fig.add_trace(go.Scatter(x = [sample_mean,sample_mean],y = [0,0.15],mode = 'lines',name = 'mean of sample data'))
fig.show()
```
